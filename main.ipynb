Importar los datos

```python
import pandas as pd
import numpy as np

# lectura del dataset
encuestaSalarial = pd.read_csv('Datos/EncuestaRemuneracionSalarialArgentina2023.csv')
print("Dataset original")
print(encuestaSalarial)

# Elimina registros duplicados
encuestaSalarial = encuestaSalarial.drop_duplicates()
print("Dataset sin duplicados")
print(encuestaSalarial)
```

Imprimir los 10 primeros registros y describir el dataset

```python
print ("Los 10 primeros registros del dateset:")
display(encuestaSalarial.head(10))

print ("Descripcion del dataset: ")
display(encuestaSalarial.describe())

```

Por cada columna:
• En las columnas numéricas analizar la cantidad de valores válidos sobre el total, media y mediana, desviación estándar, valores ausentes y mal registrados o erróneos.

```python
numero_columnas = encuestaSalarial.select_dtypes(include=[np.number]).columns

for columna in numero_columnas:
    print(f"Analisis de la columna: {columna}")

    total_valores = encuestaSalarial[columna].size
    valores_validos= encuestaSalarial[columna].count()
    porcentaje = valores_validos / total_valores * 100


    valor_medio = encuestaSalarial[columna].mean()
    mediana = encuestaSalarial[columna].median()

    desv_std = encuestaSalarial[columna].std()

    valores_ausentes = encuestaSalarial[columna].isnull().sum()

    valores_erroneos = encuestaSalarial[columna][encuestaSalarial[columna] < 0].count()

    print(f"Total de valores: {total_valores}")
    print(f"Valores válidos: {valores_validos}")
    print(f"Porsentaje de valores validos: {porcentaje:.2f}%")
    print(f"Media: {valor_medio}")
    print(f"Mediana: {mediana}")
    print(f"Desviación estándar: {desv_std}")
    print(f"Valores ausentes: {valores_ausentes}")
    print(f"Valores erróneos: {valores_erroneos}")
    print("\n")
```

En cuanto al uso de la librería MatPlotLib se propone realizar al menos 3 gráficos que guarden relación con los datos. Por ejemplo puede usar un gráfico tipo torta para reflejar los porcentajes dentro de una variable categórica.

Los gráficos deben estar correctamente etiquetados para que se pueda observar los datos que reflejan.

### Grafico 1

```python
import pandas as pd
import matplotlib.pyplot as plt

# Se carga dataset
data = pd.read_csv('Datos/EncuestaRemuneracionSalarialArgentina2023.csv')

#Grafico de Torta
dollar_payment_counts = data['pagos_en_dolares'].value_counts()
plt.figure(figsize=(8, 8))
plt.pie(dollar_payment_counts, labels=dollar_payment_counts.index, autopct='%1.1f%%', startangle=140, colors=['lightskyblue', 'lightgreen','lightcoral'])
plt.title('Distribución de Pagos en Dólares')
plt.axis('equal')
#plt.savefig('Grafico-PagoEnDolares.png')
plt.show()
```

### Grafico 2

```python
#Grafico de Barras
contract_counts = data['tipo_de_contrato'].value_counts()
plt.figure(figsize=(10, 6))
plt.bar(contract_counts.index, contract_counts.values, color='skyblue')
plt.xlabel('Tipo de Contrato')
plt.ylabel('Cantidad')
plt.title('Distribución de Tipo de Contrato')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
#plt.savefig('Grafico-TipoDeContrato.png')
plt.show()
```
